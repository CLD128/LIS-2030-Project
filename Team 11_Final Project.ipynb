## Data Sources

To address the questions, sources of data obtained include median age of death rates and death by overdose rates within Allegheny County as well as fatal encounters with police. Datasets pertaining to the median age of death and fatal accidental overdoses were obtained from the Western Pennsylvania Regional Data Center (WPRDC), an open data repository that aims to “make public information easier to find and use by hosting datasets” from institutions and organizations.<sup>2</sup> The WPRDC also maintains Allegheny County and the City of Pittsburgh’s open data portal. Data on fatal encounters with police was acquired from Mapping Police Violence, a research collaborative that collects data on police killings in the United States. 

Below sources and description of the data are provided: 

[Allegheny County Median Age at Death](https://data.wprdc.org/dataset/median-age-death): The dataset includes information about the median age of death for members located in Allegheny County by municipality. The data is based on the “deceased's residence at the time of death and not the location where the actual death occurred.”<sup>3</sup>  Median age is based on official death records released by the Pennsylvania Department of Health. Data is broken down by race and includes the count of deaths for Allegheny County Municipalities as well as neighborhoods within the City of Pittsburgh. The dataset was useful in determining the median age for both White and African American men between the years ranging from 2013 to 2015 in Allegheny County. 

[Mapping Police Violence](https://mappingpoliceviolence.org/): The data provides information on individuals who died during fatal counters with the police in the United States. The dataset was utilized to determine the race and median age of men within Allegheny County who were victims of police brutality. According to the Mapping Police Violence project, a police killing is defined as “a case where a person dies as a result of being shot, beaten, restrained, intentionally hit by a police vehicle, pepper-sprayed, tasered, or otherwise harmed by police officers, whether on-duty or off-duty.”<sup>4</sup>

[Fatal Accidental Overdoses](https://data.wprdc.org/dataset/allegheny-county-fatal-accidental-overdoses): The dataset from the Western Pennsylvania Regional Data Center provides information such as the date, zip code of incident, zip code of residence, age, gender, and race of individuals who died by drug overdose in Allegheny County. In this case, the zip code of incident is considered the location of “where the Office of the Medical Examiner received the body and not necessarily where the overdose occurred.”<sup>5</sup> (WPRDC). The data will be useful in determining rates of drug overdoses for white and African American men in Allegheny County from the years ranging from 2013 to 2015.

This next section was a bit tricky, and with assistance from Dr. Biehl, we used "astype - datetime64," which transferred the data within the column to a data/time field so we could find the incidents within the years from 2013 to 2015. (A helpful way to think about it is "Pandas dtype *datetime64*" = "Python data type = *datetime*"), just as Pandas's dtype "int64" is equal to an "integer" in regular old Python. Same kind of data, just named differently in the library.) After that, we created a query mask to find the incidents between 2013 and 2015, and a second query mask to find the incidents that took place in Allegheny County. Assigning the variable "allegheny_violence" to the combination of both query masks, we found our incidents. Rather than indexing them by numbers from the original dataset, we indexed the resulting data by the victim's name.

```python
police_data["Date of Incident (month/day/year)"] = police_data["Date of Incident (month/day/year)"].astype("datetime64[ns]")
year_county_mask = (police_data["Date of Incident (month/day/year)"] > "1-1-2013") & (police_data["Date of Incident (month/day/year)"] <= "12-31-2015")
county_query_mask = police_data["County"] == "Allegheny"
allegheny_violence = police_data[county_query_mask & year_county_mask]
allegheny_violence.set_index("Victim's name")
```

```python
wprdcxb_victims["Date of Incident (month/day/year)"] = wprdcxb_victims["Date of Incident (month/day/year)"].astype("datetime64[ns]")
black_encounter_2013 = (wprdcxb_victims["Date of Incident (month/day/year)"] > "1-1-2013") & (wprdcxb_victims["Date of Incident (month/day/year)"] <= "12-31-2013")
medianpe_2013 = wprdcxb_victims[black_encounter_2013]
black_encounter_2014 = (wprdcxb_victims["Date of Incident (month/day/year)"] > "1-1-2014") & (wprdcxb_victims["Date of Incident (month/day/year)"] <= "12-31-2014")
medianpe_2014 = wprdcxb_victims[black_encounter_2014]
black_encounter_2015 = (wprdcxb_victims["Date of Incident (month/day/year)"] > "1-1-2015") & (wprdcxb_victims["Date of Incident (month/day/year)"] <= "12-31-2015")
medianpe_2015 = wprdcxb_victims[black_encounter_2015]
```

And we do the same for overdose deaths, using the "case_year" column.

```python
white_overdose_2013 = (raceMW["case_year"] == 2013)
wmedianod_2013 = raceMW[white_overdose_2013]
white_overdose_2014 = (raceMW["case_year"] == 2014)
wmedianod_2014 = raceMW[white_overdose_2014]
white_overdose_2015 = (raceMW["case_year"] == 2015)
wmedianod_2015 = raceMW[white_overdose_2015]

black_overdose_2013 = (raceMB["case_year"] == 2013)
bmedianod_2013 = raceMB[black_overdose_2013]
black_overdose_2014 = (raceMB["case_year"] == 2014)
bmedianod_2014 = raceMB[black_overdose_2014]
black_overdose_2015 = (raceMB["case_year"] == 2015)
bmedianod_2015 = raceMB[black_overdose_2015]
```
