For this portion of the project, we set out to determine the amount of both black men and white men who died in Allegheny County from police encounters between the years of 2013 and 2015. In order to find this number from the "Mapping Police Violence" Dataset, we used Pandas data frames as shown below.

```python
import pandas as pd
```

After importing the Pandas library into Python, we went to "MappingPoliceViolence.org" and downloaded their dataset, which includes a list of all individuals in the US who died during fatal encounters with the police. After downloading this as an Excel file onto a desktop, we saved the file as a CSV file, which allowed us to upload this into Jupyter Hub. We then opened the file as a Pandas data frame, as you can see below. Please note that we are using "head" to make data frames shorter and add to the notebook's readability.

```python
police_data = pd.read_csv("files/MPV.csv")
police_data.head()
```

This next section was a bit tricky, and with assistance from Dr. Biehl, we used "astype - datetime64", which transfered the data within the column to a data/time field so we could find the incidents within the years we were looking for, 2013 to 2015. (A helpful way to think about it is "Pandas dtype *datetime64*" = "Python data type = *datetime*"), just as Pandas's dtype "int64" is equal to an "integer" in regular old Python. Same kind of data, just named differently in the library.) After that, we created a query mask to find the incidents between 2013 and 2015, and a second query mask to find the incidents that took place in Allegheny County. Assigning the variable "allegheny_violence" to the combination of both query masks, we found our incidents. Rather than indexing them by numbers from the original dataset, we indexed the resulting data by the victim's name.

```python
police_data["Date of Incident (month/day/year)"] = police_data["Date of Incident (month/day/year)"].astype("datetime64[ns]")
year_county_mask = (police_data["Date of Incident (month/day/year)"] > "1-1-2013") & (police_data["Date of Incident (month/day/year)"] <= "12-31-2015")
county_query_mask = police_data["County"] == "Allegheny"
allegheny_violence = police_data[county_query_mask & year_county_mask]
allegheny_violence.set_index("Victim's name")
```

While not exactly the number we're looking for, let's check in and see how many individuals were killed in Allegheny County during this time frame during police encounters. (The bottom of the dataset gives us a hint, but let's double check.)

```python
len(allegheny_violence)
```

There's our confirmation - 9 individuals killed. Before performing this data manipulation, we weren't sure if we were going to have to factor out the victims by gender or other races, as we are only looking at black and white men for this project. As you can see, we don't need to do that, as every individual in this new dataset was a black or white male. Now, in our request to fully answer Question 1, we need to further break down the data sets by race (black/white), as well as find the median age at time of death for these two groups of individuals. We're using another query mask to narrow down the dataset to only include victims whose race is "Black."

```python
b_mask = allegheny_violence["Victim's race"] == "Black"
b_victims = allegheny_violence[b_mask]
b_victims
```

After that, we use the Data Frame function "median" to determine the median age.

```python
b_victims["Victim's age"].median()
```

We're now going to repeat the process with white victims to find their median age.

```python
w_mask = allegheny_violence["Victim's race"] == "White"
w_victims = allegheny_violence[w_mask]
w_victims
```

```python
w_victims["Victim's age"].median()
```

We found the answers we were looking for - the median age for black men killed in police encounters in Allegheny county between 2013 and 2015 is 25, and the median age for their white counterparts in the same scenario is 27.

Now, we want to compare the ages of death for these 9 individuals with the median age of death for those in the same municipality.

Again, we start by importing the data into Python using a Pandas library on the median age of death for this period. In particular, the Western Pennsylvania Regional Data Center has tracked the median age at death in Allegheny County between 2011 and 2015 and made that available at "https://data.wprdc.org/dataset/median-age-death".

Fortunately, this already came as a CSV file, and we could then upload it to Jupyter Hub. Reading the file into Python results in the following table showing 131 Allegheny County municipalities and their respective data.

```python
import pandas as pd
wprdc_data = pd.read_csv("files/munimedianage20112015.csv")
wprdc_data.head()
```

Looking through the municipalities here and comparing to our list of shooting victims, we can see we will have a few issues merging these datasets. First, there's a pesky misspelling of "Pittsburg" in Shawn Evans' listing. Let's fix that now with a quick "replace" method.

```python
w_victims_sic = w_victims.replace("Pittsburg","Pittsburgh")
w_victims_sic[w_victims_sic["Victim's name"] == "Shawn Evans"]
```

Another issue is that the WPRDC data does not include a listing for the Knoxville neighborhood. However, the WPRDC data does include Mount Oliver, which is adjacent to Knoxville, and so we will substitute Knoxville into the WPRDC dataframe to get the median age of death for the adjacent neighborhood.

```python
wprdc_sub = wprdc_data.replace("Mount Oliver Borough","Knoxville")
wprdc_sub[wprdc_sub.Municipality == "Knoxville"]
```

The last, more serious issue, is that the "City" designation in our table does not include Township or Borough identifiers that the WPRDC data does. We drop those designations and merge each table. After merging the data, we want to highlight the important information. So, we create a new column right after "Victim's Age" to input the median age at death in our dataset, and we create another new column to show the median age at death in their neighborhood. This will help us make a graphic later on.

```python
wprdc_sub["City"] = wprdc_sub["Municipality"].str.replace("\sBorough|\sTownship","")
wprdc_sub
```

```python
wprdcxw_victims = w_victims_sic.merge(wprdc_sub)
wprdcxw_victims.insert(2,"Median Age at Death in Neighborhood",wprdcxw_victims["WHITE MD AGE AT DEATH"])
wprdcxw_victims.insert(3,"Median Age at Death",w_victims["Victim's age"].median())
wprdcxw_victims
```

```python
wprdcxb_victims = b_victims.merge(wprdc_sub)
wprdcxb_victims.insert(2,"Median Age at Death in Neighborhood",wprdcxb_victims["BLACK MD AGE AT DEATH"])
wprdcxb_victims.insert(3,"Median Age at Death",b_victims["Victim's age"].median())
wprdcxb_victims
```

Now, we can get an answer to which demographic group (white or black men) die from overdoses or police encounters closest to their median age. Since our data cannot give a direct answer to median age at death in Allegheny County, we will take the average of the medians from the Western Pennsylvania Regional Data Center data. It seems some municipalities didn't report ages of death for black people, so we will keep those out of the average.

```python
bvictim_pe_median = wprdcxb_victims["Victim's age"].median()
ballegheny_avg = round(wprdc_data["BLACK MD AGE AT DEATH"][59:129].mean(),2)
difference1 = round(ballegheny_avg - bvictim_pe_median,2)
print("The median age for black males to die from a police encounter between 2013 and 2015 was: " + str(bvictim_pe_median) + ", which was " + str(difference1) + " years less than the average age at death for black people in Allegheny County of " + str(ballegheny_avg) + " years.")
```

```python
wvictim_pe_median = wprdcxw_victims["Victim's age"].median()
wallegheny_avg = wprdc_data["WHITE MD AGE AT DEATH"].mean()
difference2 = round(wallegheny_avg - wvictim_pe_median, 2)
print("The median age for white males to die from a police encounter between 2013 and 2015 was: " + str(wvictim_pe_median) + ", which was " + str(difference2) + " years less than the average age at death for white people in Allegheny County of " + str(wallegheny_avg) + " years.")
```

```python
b_od_median = raceMB['age'].median()
difference3 = round(ballegheny_avg - b_od_median,2)
print("The median age for black males to die from an overdose between 2013 and 2015 was: " + str(b_od_median) + ", which was " + str(difference3) + " years less than the average age at death for black people in Allegheny County of " + str(ballegheny_avg) + " years.")
```

```python
w_od_median = raceMW['age'].median()
difference4 = round(wallegheny_avg - w_od_median,2)
print("The median age for white males to die from an overdose between 2013 and 2015 was: " + str(w_od_median) + ", which was " + str(difference4) + " years less than the average age at death for black people in Allegheny County of " + str(wallegheny_avg) + " years.")
```

This shows that black men dying of overdoses is the closest demographic to the average in Allegheny County, losing over 17 years compared to their counterparts in the community.

Next, we want to break down the median age of deaths, split by race, for both police encounters and overdose deaths in each year. For this, we need matplotlib. But first, we need the median ages split by year. This is easiest for white victims of police encounters since they all occurred in 2013, but for the rest, we need to use a variation on our earlier masks.

```python
wprdcxb_victims["Date of Incident (month/day/year)"] = wprdcxb_victims["Date of Incident (month/day/year)"].astype("datetime64[ns]")
black_encounter_2013 = (wprdcxb_victims["Date of Incident (month/day/year)"] > "1-1-2013") & (wprdcxb_victims["Date of Incident (month/day/year)"] <= "12-31-2013")
medianpe_2013 = wprdcxb_victims[black_encounter_2013]
black_encounter_2014 = (wprdcxb_victims["Date of Incident (month/day/year)"] > "1-1-2014") & (wprdcxb_victims["Date of Incident (month/day/year)"] <= "12-31-2014")
medianpe_2014 = wprdcxb_victims[black_encounter_2014]
black_encounter_2015 = (wprdcxb_victims["Date of Incident (month/day/year)"] > "1-1-2015") & (wprdcxb_victims["Date of Incident (month/day/year)"] <= "12-31-2015")
medianpe_2015 = wprdcxb_victims[black_encounter_2015]
```

And we do the same for overdose deaths, using the "case_year" column.

```python
white_overdose_2013 = (raceMW["case_year"] == 2013)
wmedianod_2013 = raceMW[white_overdose_2013]
white_overdose_2014 = (raceMW["case_year"] == 2014)
wmedianod_2014 = raceMW[white_overdose_2014]
white_overdose_2015 = (raceMW["case_year"] == 2015)
wmedianod_2015 = raceMW[white_overdose_2015]

black_overdose_2013 = (raceMB["case_year"] == 2013)
bmedianod_2013 = raceMB[black_overdose_2013]
black_overdose_2014 = (raceMB["case_year"] == 2014)
bmedianod_2014 = raceMB[black_overdose_2014]
black_overdose_2015 = (raceMB["case_year"] == 2015)
bmedianod_2015 = raceMB[black_overdose_2015]
```

Ok, now that we split our data by year, we can calculate the median directly into our dataset and make a nice, neat graph!

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['2013', '2014', '2015']
black_encounter = [medianpe_2013["Victim's age"].median(), medianpe_2014["Victim's age"].median(), medianpe_2015["Victim's age"].median()]
white_encounter = [wprdcxw_victims["Victim's age"].median(), 0, 0]
black_overdose = [bmedianod_2013["age"].median(), bmedianod_2014["age"].median(), bmedianod_2015["age"].median()]
white_overdose = [wmedianod_2013["age"].median(), wmedianod_2014["age"].median(), wmedianod_2015["age"].median()]

x = np.arange(len(labels))
width = 0.3

fig, ax = plt.subplots()
rects1 = ax.bar(x - 1.5*width, white_encounter, width, label='Black Encounter')
rects2 = ax.bar(x - width/2, black_encounter, width, label='White Encounter')
rects3 = ax.bar(x + width/2, white_overdose, width, label='Black Overdose')
rects4 = ax.bar(x + 1.5*width, black_overdose, width, label='White Overdose')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Median Age at Death')
ax.set_title('Comparison of Median Ages: Police Encounters / Overdoses')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc=2, fontsize='small')


fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots()

x = np.linspace(0,2,3)
black_encounter = [medianpe_2013["Victim's age"].median(), medianpe_2014["Victim's age"].median(), medianpe_2015["Victim's age"].median()]
line1 = ax.plot(x, black_encounter, label='Black Encounter')

x2 = np.linspace(0,2,3)
white_encounter = [wprdcxw_victims["Victim's age"].median(), 0, 0]
line1 = ax.plot(x2, white_encounter, label='White Encounter')

x3 = np.linspace(0,2,3)
black_overdose = [bmedianod_2013["age"].median(), bmedianod_2014["age"].median(), bmedianod_2015["age"].median()]
line1 = ax.plot(x3, black_overdose, label='Black OD')

x4 = np.linspace(0,2,3)
white_overdose = [wmedianod_2013["age"].median(), wmedianod_2014["age"].median(), wmedianod_2015["age"].median()]
line1 = ax.plot(x4, white_overdose, label='White OD')
labels = ['2013', '2014', '2015']
plt.title('Comparison of Median Ages: Police Encounters / Overdoses')
labels = ["2013", "2014", "2015"]
plt.xticks(np.arange(3), labels)
plt.ylabel("Median Age at Death")
ax.legend()
plt.show()
```

As we can see, while the median age of police encounter deaths for both genders is generally less than those for overdoses, there is a lot more variability in that data since we are dealing with less than 5 individuals in each category per year.

For the next chart, we look at the difference between deaths, split by race, between overdose deaths and police encounter deaths to examine the scope of each problem from 2013 to 2015.

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['Black Encounter', 'White Encounter', 'Black OD', 'White OD']
deaths = [len(b_victims["Victim's age"]), len(w_victims["Victim's age"]), len(raceMB['age']), len(raceMW['age'])]

x = np.arange(len(labels))
width = 0.35


fig, ax = plt.subplots()
ax.bar(labels, deaths, color = "black")
ax.set_title('Comparison of Total Deaths: Police Encounters / Overdoses')
ax.set_ylabel('Deaths (2013-2015)')
ax.set_yscale('log')
ax.set_xticks(x)
ax.set_xticklabels(labels)

fig.tight_layout()


plt.show()
```

We had to use a logarithmic scale on the y-axis here because otherwise the "encounter" bars were barely visible lines. Still, as we can see, overdose deaths far outstripped fatalities from police encounters. While white men were more likely to overdose between 2013 and 2015, black men were more likely to die at the hands of police.
